### Un Problema de Minimización en Prospección Sismica 

Descripción general de la prospección

· Este problema aborda cómo se pueden investigar las estructuras subterraneas utilizando ondas artificiales producidas en el suelo. El método consiste en analizar el tiempo que las ondas tardan e viajar desde una fuente hasta un receptor a través de diferentes capas de medios.

· El problema involucra dos tipos de ondas: 

1. _Onda de Superficie_: Se desplaza sobre la superficie del suelo a velocidad $v_1$. El tiempo que tarda en recorrer una distancia $s$ está dado por $t_1=\frac{s}{v_1}$
2. _Onda que viaja a través de las capas_: Esta onda penetra en el suelo y se mueve por diferentes medios con velocidades $v_1$ y $v_2$. El tiempo que toma en atravesar ambas capas está dado por $t_2$, que depende la geometría del sistema.

· Problema de Minimización:

Para minimizar el tiempo $t_2$, se busca el ángulo óptimo $\theta_1$, que describe el trayecto de la onda en el primer medio. A partir de la condición de minimización, se obtiene una relación entre $\theta_1$ y las velocidades $v_1$ y $v_2$. Se propone un método gráfico para determinar las velocidades $v_2$ y el grosor de la capa $d_1$ mediante un análisis experimental.

· Superficie Inclinada: 

En casos donde la interfaz entre dos medios está inclinada o tiene una forma desconocidad, como en la exploración petrolera, es posible determinar su geometría comparando los resultados experimentales con predicciones teóricas. Para estructuras más complejas, como una cúpula enterrada, se puede extender el analisis a dos dimensiones y considerar multiples posiciones de la fuente y el receptor.

· Caso de Tres Medios:

El texto también explora un caso donde la onda viaja a través de tres medios con diferentes velocidades. En este caso, el tiempo total de viaje $t_2$ se calcula sumando los tiempo en cada medio, lo que introduce una mayor complejidad en el problema de minimización.

### Problema de minimización en prospección sísmica.

En prospección sísmica, se investigan las estructuras subterráneas mediante ondas artificialmente generadas en el suelo. Supón que existe una capa de espesor uniforme $d_1$, con velocidad de onda constante $v_1$, y sobre ella se encuentra una segunda capa con velocidad constante $v_2$, donde $v_2>v_1$. Un dispositivo en la fuente genera un pulso de onda, y un sistema de recepción registra los tiempos que tarda dicho pulso en llegar desde la funete hasta varios puntos en la superficie de la Tierra. Se conocen las siguientes condiciones:

1. _Onda de superficie_: se desplaza directamente desde la fuente hasta el receptor a lo largo de la superficie con velocidad $v_1$. El tiempo que tarda está dado por

   $t_1=\frac{s}{v_1}$

   donde $s$ es la distancia entre la fuente y el receptor.
2. _Onda reflejada_: la onda viaja hacia la interfaz entre los dos medios, se propaga por el medio inferior con velocidad $v_2$ y luego regresa a traves del medio superior. El tiempo total que toma esta onda está dado por:
   
   $t_2=\frac{2d_1}{v_1}sec\theta_1+\frac{s-2d_1tan\theta_1}{v_2}$

    donde $\theta_1$ es el ángulo que la trayectoria de la onda forma con la vertical.

   

Preguntas:

a) Demuestrar que el valor de  $\theta_1$  que minimiza el tiempo total $t_2$ satisface la ecuación  $\sin \theta_1 = \frac{v_1}{v_2}$. 

b) Introducir el parámetro $k = \frac{v_1}{v_2}$ y simplifica la ecuación para $t_2$.

c) Suponiendo que se han medido experimentalmente los valores de $t_1$ y $t_2$ para varios valores de $s$, graficar  $y = \frac{t_2}{t_1}$  contra  $x = \frac{1}{s}$ . Demuestrar que la intersección de la línea con el eje $y$ da el valor de $k$, y que la pendiente de la línea permite encontrar el espesor $d_1$ de la capa superior.



### a) Demuestrar que el valor de $ \theta_1 $ que minimiza el tiempo total $t_2$ satisface la ecuación  $\sin \theta_1 = \frac{v_1}{v_2}$.

Para minimizar el tiempo $t_2$, tomamos la derivada de $t_2$ con respecto a $ \theta_1 $ y la igualamos a cero:

$
t_2 = \frac{2d_1}{v_1} \sec \theta_1 + \frac{s - 2d_1 \tan \theta_1}{v_2}
$

Derivando con respecto a $ \theta_1 $:

$
\frac{dt_2}{d\theta_1} = \frac{2d_1}{v_1} \frac{d}{d\theta_1}(\sec \theta_1) + \frac{-2d_1}{v_2} \frac{d}{d\theta_1}(\tan \theta_1)
$

Sabemos que:

$
\frac{d}{d\theta_1}(\sec \theta_1) = \sec \theta_1 \tan \theta_1 \quad \text{y} \quad \frac{d}{d\theta_1}(\tan \theta_1) = \sec^2 \theta_1
$

Sustituyendo:

$
\frac{dt_2}{d\theta_1} = \frac{2d_1 \sec \theta_1 \tan \theta_1}{v_1} - \frac{2d_1 \sec^2 \theta_1}{v_2}
$

Igualamos a cero para encontrar el valor de $ \theta_1 $ que minimiza $t_2$:

$
\frac{2d_1 \sec \theta_1 \tan \theta_1}{v_1} = \frac{2d_1 \sec^2 \theta_1}{v_2}
$

Cancelamos $2d_1$ y $ \sec \theta_1 $:

$
\frac{\tan \theta_1}{v_1} = \frac{\sec \theta_1}{v_2}
$

Recordemos que $ \tan \theta_1 = \frac{\sin \theta_1}{\cos \theta_1}$ y $ \sec \theta_1 = \frac{1}{\cos \theta_1} $, por lo que la ecuación se convierte en:

$
\frac{\sin \theta_1}{v_1 \cos \theta_1} = \frac{1}{v_2 \cos \theta_1}
$

Cancelamos  $\cos \theta_1 $ y obtenemos:

$
\sin \theta_1 = \frac{v_1}{v_2}
$

Este es el valor de $ \theta_1 $ que minimiza $t_2$, como se pedía demostrar.

---

### b) Introducir el parámetro $k = \frac{v_1}{v_2}$ y simplifica la ecuación para $t_2$.

Partimos de la ecuación para $t_2$:

$
t_2 = \frac{2d_1}{v_1} \sec \theta_1 + \frac{s - 2d_1 \tan \theta_1}{v_2}
$

Sabemos que  $\sin \theta_1 = \frac{v_1}{v_2} $, por lo que $\cos \theta_1 = \sqrt{1 - \sin^2 \theta_1} = \sqrt{1 - \left(\frac{v_1}{v_2}\right)^2} $. Esto nos lleva a:

$
\sec \theta_1 = \frac{1}{\cos \theta_1} = \frac{1}{\sqrt{1 - \left(\frac{v_1}{v_2}\right)^2}} = \frac{1}{\sqrt{1 - k^2}}
$

También, $ \tan \theta_1 = \frac{\sin \theta_1}{\cos \theta_1} = \frac{\frac{v_1}{v_2}}{\sqrt{1 - \left(\frac{v_1}{v_2}\right)^2}} = \frac{k}{\sqrt{1 - k^2}} $.

Sustituyendo estos valores en la ecuación para $t_2$, obtenemos:

$
t_2 = \frac{2d_1}{v_1} \frac{1}{\sqrt{1 - k^2}} + \frac{s - 2d_1 \frac{k}{\sqrt{1 - k^2}}}{v_2}
$

Simplificando:

$
t_2 = \frac{2d_1}{v_1} \frac{1}{\sqrt{1 - k^2}} + \frac{s}{v_2} - \frac{2d_1 k}{v_2 \sqrt{1 - k^2}}
$

Agrupando términos:

$
t_2 = \frac{2d_1}{v_1} \frac{1}{\sqrt{1 - k^2}} + \frac{s}{v_2} - \frac{2d_1 k}{v_2 \sqrt{1 - k^2}}
$

Ahora, podemos combinar los términos que contienen $ \frac{1}{\sqrt{1 - k^2}} $ en un solo término. Reescribiendo:

$
t_2 = \frac{2d_1}{v_1} \frac{1}{\sqrt{1 - k^2}} + \left( \frac{s}{v_2} - \frac{2d_1 k}{v_2 \sqrt{1 - k^2}} \right)
$

Para simplificar aún más, observamos que el parámetro $ k = \frac{v_1}{v_2} $, lo que nos permite agrupar los términos de manera más clara, pero la ecuación no se simplifica significativamente más sin conocer valores específicos de los parámetros $d_1$, $v_1$, $v_2$ y $s$.

Finalmente, $t_2$ está ahora expresado en términos del parámetro $k = \frac{v_1}{v_2}$, que era lo que se requería.

---

### c) Suponiendo que se han medido experimentalmente los valores de $t_1$ y $t_2$ para varios valores de $s$, grafica $y = \frac{t_2}{t_1} $ contra $ x = \frac{1}{s} $. Demuestra que la intersección de la línea con el eje $y$ da el valor de $k$, y que la pendiente de la línea permite encontrar el espesor $d_1$ de la capa superior.

Sabemos que:

- $ t_1 = \frac{s}{v_1} $
- $ t_2 = \frac{2d_1}{v_1} \frac{1}{\sqrt{1 - k^2}} + \frac{s}{v_2} - \frac{2d_1 k}{v_2 \sqrt{1 - k^2}} $

Dividimos $ t_2 $ entre $ t_1 $:

$
\frac{t_2}{t_1} = \frac{\left( \frac{2d_1}{v_1} \frac{1}{\sqrt{1 - k^2}} + \frac{s}{v_2} - \frac{2d_1 k}{v_2 \sqrt{1 - k^2}} \right)}{\frac{s}{v_1}}
$

Simplificando esta expresión, obtenemos:

$
\frac{t_2}{t_1} = \frac{2d_1}{s \sqrt{1 - k^2}} + k
$

Esta es una ecuación lineal de la forma \( y = mx + b \), donde:

- $ y = \frac{t_2}{t_1} $
- $ x = \frac{1}{s} $
- $ b = k = \frac{v_1}{v_2} $, que es la intersección con el eje $y$.
- La pendiente $m$ es  $\frac{2d_1}{\sqrt{1 - k^2}} $.

Por lo tanto, al graficar $ y $ contra $ x $, la intersección con el eje $y$ nos da el valor de $k$, que a su vez permite calcular la relación entre las velocidades $ v_1 $ y $ v_2 $.

Además, la pendiente $m$ de la línea recta permite determinar el espesor $d_1$ de la capa superior a partir de la relación:

$
d_1 = \frac{m \sqrt{1 - k^2}}{2}
$

Este método experimental nos permite estimar tanto $d_1$ como $k = \frac{v_1}{v_2}$, utilizando los datos obtenidos de $t_1$ y $t_2$.



# Caso de Superficie Inclinada

En ciertas aplicaciones de prospección sísmica, la interfaz entre dos capas de suelo no es plana, sino inclinada. Considera una capa de espesor $d_1$ con velocidad de onda $v_1$ y una segunda capa con velocidad $v_2$, tal que $v_2 > v_1$. La interfaz entre las dos capas está inclinada, como se muestra en la figura 4.7.



Supón que se han tomado medidas de los tiempos $t_1$ y $t_2$ para varias posiciones de la fuente $A$, $B$ y $C$, y receptores $P$, $Q$ y $R$, ubicados en diferentes posiciones a lo largo de la superficie inclinada.

#### Preguntas:

a) **Derivación de la trayectoria mínima**:

   Dado que la interfaz entre las capas está inclinada, la onda reflejada no seguirá la misma trayectoria que en el caso de una interfaz plana. Escribe la expresión para el tiempo total $t_2$ que la onda reflejada tarda en viajar desde la fuente hasta el receptor. Considera que la onda deberá viajar en dos direcciones inclinadas.

b) **Condiciones de minimización**:

   Deriva la condición que debe satisfacer el ángulo de incidencia $\theta_1$ para minimizar el tiempo total $t_2$. Compara esta condición con la obtenida para el caso de la interfaz plana. ¿Cómo afecta la inclinación de la interfaz a la trayectoria óptima de la onda?

c) **Interpretación experimental**:

   Explica cómo se podría verificar experimentalmente si la interfaz entre dos capas está inclinada utilizando las mediciones de $t_1$ y $t_2$ en diferentes puntos de la superficie. ¿Qué patrones esperarías observar en los tiempos medidos si la interfaz estuviera inclinada?

---



### Respuestas 

**a) Derivación de la trayectoria mínima:**

Cuando la interfaz entre las dos capas está inclinada, la onda reflejada no sigue la misma trayectoria que en el caso de una interfaz plana. El tiempo total $t_2$ para la onda reflejada se puede modificar para considerar la inclinación de la interfaz:

$
t_2 = \frac{2d_1 \sec \theta_1}{v_1} + \frac{s - 2d_1 \tan \theta_1}{v_2}
$

Aquí, $d_1$ representa la distancia perpendicular a la interfaz inclinada, y la trayectoria de la onda reflejada cambia con respecto a la inclinación. Si la superficie está inclinada, las longitudes de las trayectorias $s_1$ y $s_2$ entre la fuente y el receptor cambian de acuerdo a la inclinación.

**b) Condiciones de minimización:**

Para encontrar el valor de $ \theta_1 $ que minimiza el tiempo total $t_2$, diferenciamos la ecuación con respecto a $ \theta_1 $ y establecemos la derivada igual a cero. La condición de minimización es similar a la obtenida para el caso de una interfaz plana, pero con el efecto añadido de la inclinación:

$
\frac{dt_2}{d\theta_1} = \frac{2d_1 \sin \theta_1}{v_1 \cos^2 \theta_1} - \frac{2d_1}{v_2 \cos^2 \theta_1} = 0
$

Esto lleva a la misma relación que en el caso de la interfaz plana:

$
\sin \theta_1 = \frac{v_1}{v_2}
$

La inclinación afecta el valor de $s$, pero la condición para el ángulo $ \theta_1 $ que minimiza $t_2$ no cambia. Sin embargo, la trayectoria de la onda sí se ve alterada.

**c) Interpretación experimental:**

Si la interfaz entre las capas está inclinada, los tiempos $t_1$ y $t_2$ medidos para diferentes posiciones de la fuente y el receptor no seguirán una relación lineal como en el caso de la interfaz plana. Específicamente, los tiempos medidos para posiciones diferentes de la fuente $A$, $B$, $C$ y receptores $P$, $Q$, $R$ mostrarán variaciones que reflejan la inclinación de la interfaz.

Un patrón esperado es que los tiempos $t_2$ serán mayores o menores dependiendo de si el receptor está más cerca o más lejos de la inclinación. Si la inclinación es pronunciada, los tiempos de recorrido reflejados desde la interfaz inclinada serán diferentes en comparación con una interfaz plana.

### Caso de Tres Medios

Ahora consideramos un caso más complejo en el que hay **tres medios** con diferentes velocidades de onda. Las velocidades en cada medio son $v_1$, $v_2$, y $v_3$, con $v_3 > v_2 > v_1$. La onda viaja a través de las tres capas, como se muestra en la figura 4.8. El espesor de la primera capa es $d_1$ y el de la segunda es $d_2$.



El tiempo total $t_2$ que la onda tarda en viajar desde la fuente hasta el receptor está dado por:

$
t_2 = \frac{2d_1}{v_1} \sec \theta_1 + \frac{2d_2}{v_2} \sec \theta_2 + \frac{s - 2d_1 \tan \theta_1 - 2d_2 \tan \theta_2}{v_3}
$

#### Preguntas:

a) **Minimización del tiempo total**:

   Encuentra los valores de $\theta_1 $ y $ \theta_2 $ que minimizan el tiempo total $t_2$. Utiliza las condiciones de minimización para derivar las siguientes ecuaciones:
   
$
   \sin \theta_1 = \frac{v_1}{v_3}, \quad \sin \theta_2 = \frac{v_2}{v_3}
$

b) **Simplificación de la ecuación para $t_2$**:

   Introduce los parámetros $k_1 = \frac{v_1}{v_3}$ y $k_2 = \frac{v_2}{v_3}$. Simplifica la expresión para $t_2$ en función de estos parámetros. 

c) **Interpretación de los resultados**:

   Dado que en este caso hay tres medios involucrados, ¿cómo cambia la gráfica de $ \frac{t_2}{t_1} $contra $ \frac{1}{s} $ en comparación con el caso de dos medios? ¿Cómo puedes determinar los espesores $d_1$ y $d_2$ a partir de los resultados experimentales?

---

Estas preguntas están diseñadas para guiar al estudiante en la resolución de los casos más complejos, enfocándose en el análisis matemático y en cómo se puede interpretar experimentalmente los resultados para determinar características del subsuelo.

### Respuestas para el Caso de Tres Medios

**a) Minimización del tiempo total:**

Para encontrar los valores de $ \theta_1 $ y $ \theta_2 $ que minimizan el tiempo total $t_2$, derivamos la ecuación del tiempo total con respecto a $ \theta_1 $ y $ \theta_2 $:

$
t_2 = \frac{2d_1 \sec \theta_1}{v_1} + \frac{2d_2 \sec \theta_2}{v_2} + \frac{s - 2d_1 \tan \theta_1 - 2d_2 \tan \theta_2}{v_3}
$

Diferenciando $t_2$ con respecto a $ \theta_1 $ y $ \theta_2 $, y estableciendo las derivadas iguales a cero, obtenemos las siguientes ecuaciones para $ \theta_1 $ y $ \theta_2 $:

$
\frac{dt_2}{d\theta_1} = \frac{2d_1 \sin \theta_1}{v_1 \cos^2 \theta_1} - \frac{2d_1}{v_3 \cos^2 \theta_1} = 0
$

$
\frac{dt_2}{d\theta_2} = \frac{2d_2 \sin \theta_2}{v_2 \cos^2 \theta_2} - \frac{2d_2}{v_3 \cos^2 \theta_2} = 0
$

Esto nos da las relaciones:

$
\sin \theta_1 = \frac{v_1}{v_3}, \quad \sin \theta_2 = \frac{v_2}{v_3}
$

Estas relaciones son análogas a la ley de Snell para la refracción de ondas. Determinan los ángulos de incidencia y refracción para las ondas al pasar de un medio a otro, minimizando así el tiempo total de recorrido.

**b) Simplificación de la ecuación para \(t_2\):**

Sustituyendo $ \sin \theta_1 = \frac{v_1}{v_3} $ y $ \sin \theta_2 = \frac{v_2}{v_3} $ en la ecuación de $t_2$, podemos introducir los parámetros $ k_1 = \frac{v_1}{v_3} $ y $ k_2 = \frac{v_2}{v_3} $. Esto nos da la expresión simplificada para el tiempo total:

$
t_2 = \frac{2d_1}{v_1} (1 - k_1^2)^{1/2} + \frac{2d_2}{v_2} (1 - k_2^2)^{1/2} + \frac{s}{v_3}
$

Esta fórmula simplificada expresa el tiempo total en términos de los espesores $d_1$, $d_2$, y las velocidades de onda $v_1$, $v_2$, y $v_3$.

**c) Interpretación de los resultados:**

En el caso de tres medios, la gráfica de $ \frac{t_2}{t_1}$ contra $ \frac{1}{s} $ será más compleja que en el caso de dos medios. A medida que la onda pasa por tres capas diferentes, la relación entre $t_2$ y $t_1$ dependerá de los espesores $d_1$ y $d_2$, así como de las velocidades $v_1$, $v_2$ y $v_3$. 

La pendiente de la gráfica puede no ser lineal, y se pueden esperar diferentes segmentos de la curva para las distintas capas. A partir de los datos experimentales, es posible estimar tanto $d_1$ como $d_2$ mediante el ajuste de la gráfica a la ecuación obtenida para $t_2$, y al comparar las diferentes pendientes y curvaturas de los segmentos del gráfico.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Función para el problema sencillo
def problema_sencillo(v1, v2, d1):
    s = np.linspace(1, 20, 100)  # distancia s
    t1 = s / v1  # tiempo de onda superficial
    k = v1 / v2  # relación de velocidades
    t2 = (2 * d1 / v1) * np.sqrt(1 - k**2) + s / v2  # tiempo de onda reflejada
    
    plt.figure(figsize=(8, 6))
    plt.plot(s, t1, label="Onda superficial (t1)")
    plt.plot(s, t2, label="Onda reflejada (t2)")
    plt.xlabel("Distancia (s)")
    plt.ylabel("Tiempo (t)")
    plt.title("Caso Sencillo: Onda Superficial vs Reflejada")
    plt.legend()
    plt.grid(True)
    plt.show()

# Función para el caso de superficie inclinada
def superficie_inclinada(v1, v2, d1, inclinacion):
    s = np.linspace(1, 20, 100)  # distancia s
    t1 = s / v1  # tiempo de onda superficial
    k = v1 / v2  # relación de velocidades
    inclinacion_rad = np.deg2rad(inclinacion)  # convertir inclinación a radianes
    t2 = (2 * d1 / v1) * np.sqrt(1 - k**2) * np.cos(inclinacion_rad) + s / v2  # tiempo con inclinación
    
    plt.figure(figsize=(8, 6))
    plt.plot(s, t1, label="Onda superficial (t1)")
    plt.plot(s, t2, label="Onda reflejada con inclinación (t2)")
    plt.xlabel("Distancia (s)")
    plt.ylabel("Tiempo (t)")
    plt.title(f"Superficie Inclinada: Inclinación = {inclinacion} grados")
    plt.legend()
    plt.grid(True)
    plt.show()

# Función para el caso de tres medios
def tres_medios(v1, v2, v3, d1, d2):
    s = np.linspace(1, 20, 100)  # distancia s
    k1 = v1 / v3  # relación de velocidades para el primer medio
    k2 = v2 / v3  # relación de velocidades para el segundo medio
    t2 = (2 * d1 / v1) * np.sqrt(1 - k1**2) + (2 * d2 / v2) * np.sqrt(1 - k2**2) + s / v3  # tiempo de onda reflejada
    
    plt.figure(figsize=(8, 6))
    plt.plot(s, t2, label="Onda a través de tres medios (t2)")
    plt.xlabel("Distancia (s)")
    plt.ylabel("Tiempo (t)")
    plt.title("Caso de Tres Medios: Onda Reflejada")
    plt.legend()
    plt.grid(True)
    plt.show()

# Widgets interactivos para ajustar los parámetros
def interactivo_problema_sencillo():
    interact(problema_sencillo, 
             v1=FloatSlider(value=3, min=1, max=10, step=0.1, description='Velocidad v1'), 
             v2=FloatSlider(value=6, min=1, max=20, step=0.1, description='Velocidad v2'),
             d1=FloatSlider(value=5, min=1, max=10, step=0.1, description='Espesor d1'))

def interactivo_superficie_inclinada():
    interact(superficie_inclinada, 
             v1=FloatSlider(value=3, min=1, max=10, step=0.1, description='Velocidad v1'), 
             v2=FloatSlider(value=6, min=1, max=20, step=0.1, description='Velocidad v2'),
             d1=FloatSlider(value=5, min=1, max=10, step=0.1, description='Espesor d1'),
             inclinacion=FloatSlider(value=30, min=0, max=90, step=1, description='Inclinación'))

def interactivo_tres_medios():
    interact(tres_medios, 
             v1=FloatSlider(value=3, min=1, max=10, step=0.1, description='Velocidad v1'), 
             v2=FloatSlider(value=6, min=1, max=20, step=0.1, description='Velocidad v2'),
             v3=FloatSlider(value=9, min=1, max=30, step=0.1, description='Velocidad v3'),
             d1=FloatSlider(value=5, min=1, max=10, step=0.1, description='Espesor d1'),
             d2=FloatSlider(value=7, min=1, max=10, step=0.1, description='Espesor d2'))

# Llamamos a cada función interactiva
print("Problema Sencillo:")
interactivo_problema_sencillo()

print("Superficie Inclinada:")
interactivo_superficie_inclinada()

print("Caso de Tres Medios:")
interactivo_tres_medios()
```
